## 面向对象方法一，先把子图摆好，再按位置

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.random.randn(1000)
x=x.cumsum()
```

```python
fig, ax=plt.subplots(2,2,figsize=(10,8))
ax[0][0].plot(x,linestyle='--')
ax[0][1].hist(x,bins=50)
ax[1][0].pie(np.abs(x[:5]))
ax[1][0].axis('equal')
```

## 面向对象方法二，先设置画布，再逐一添加子图

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.random.randn(1000)
x=x.cumsum()
```

```python
fig = plt.figure(figsize=(10,8))
ax1=fig.add_subplot(2,2,1)#子图表示位置的方法和切片不同
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,1,2)
ax1.hist(x,bins=60)
ax2.pie(np.abs(x[:5]))
ax2.axis('equal')
ax3.plot(x,ls='--')
```

## 面向对象绘图的设置

# 边框和水平线

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(-np.pi,np.pi,256,endpoint = True)#在指定的间隔内返回均匀间隔的数字
```

```python
y = np.sin(x)
```

spines()对边框的设置

set_position(('data',0))对边框设置位置

ax.axhline(1,)设置水平线

```python
fig,ax=plt.subplots()
ax.plot(x,y)
ax.spines['top'].set_visible(False)#设置可见
ax.spines['right'].set_visible(False)
#ax.spines['bottom'].set_linewidth(6)
#ax.spines['left'].set_color('r')#对边框的设置
ax.spines['left'].set_position(('data',0))
ax.axhline(1,color='r',lw=2)#第一个参数代表水平的位置，hline
ax.axvline(1,color='r',lw=2)#vline垂直的线
```

## 共享x轴，y轴数据，调正子图间距

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = [1,2,3,4]
y = [4,5,8,9]
y1 = [3,6,2,5]
```

方法1：sharex,sharey共享x轴，共享y轴

```python
fig, ax=plt.subplots(2,1,sharex=True,sharey=True)
ax[0].bar(x,y)
ax[0].bar(x,y1,bottom=y)
ax[1].plot(x,y)
```

方法2：另一种初始化的共享x轴，y轴

**plt.sept(ax1.get_xticklabels(),visible=False) sept迭代，使第一张图的标签去掉**

```python
fig=plt.figure(figsize=(8,6))
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(212,sharex=ax1,sharey=ax1)#共享x轴，y轴
ax1.plot(x,y)
ax2.bar(x,y1)
plt.setp(ax1.get_xticklabels(),visible=False)
```

```python
fig=plt.figure(figsize=(8,6))
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(212,sharex=ax1,sharey=ax1)#共享x轴，y轴
ax1.plot(x,y)
ax2.bar(x,y1)
for label in ax1.xaxis.get_ticklabels():#获得x轴刻度
    label.set_visible(False)
```

## 调整子图间距，保存图像

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = [1,2,3,4]
y = [4,5,8,9]
```

plt.tight_layout()自动调间距大小  plt.subplots_adjust()自定义调整

```python
fig,ax = plt.subplots(2,2,figsize = (8,8))
ax[0,0].bar(x,y)
ax[0,1].barh(x,y)
ax[1][0].plot(x,y)
ax[1][1].scatter(x,y)
plt.subplots_adjust(wspace=0.3,hspace=0.3)
```

```python
fig,ax = plt.subplots(2,2,figsize = (8,8))
ax[0,0].bar(x,y)
ax[0,1].barh(x,y)
ax[1][0].plot(x,y)
ax[1][1].scatter(x,y)
plt.tight_layout()
```

## 保存图像

plt.savefig('图像.格式')可以控制分别率，保存格式

支持的格式 Most backends support png, pdf, ps, eps and svg.

```python
fig,ax = plt.subplots(2,2,figsize = (8,8))
ax[0,0].bar(x,y)
ax[0,1].barh(x,y)
ax[1][0].plot(x,y)
ax[1][1].scatter(x,y)
plt.subplots_adjust(wspace=1,hspace=0.5)
plt.savefig('D:/面向对象绘图073.png')#如果没用魔术命令，需要先保存再plt.show
```
