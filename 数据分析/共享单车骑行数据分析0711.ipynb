```python
data.tail()#好多Nan值
```

```python
data=data[data['StartDate']==data['EndDate']]#给data更新值
```

思考要取哪些值

```python
data = data[['MemberType','TripDurationSec','StartDay','StartTime','StartDate','StartStationName','EndStationName']]
```

```python
data=data.dropna(how='all')#去除空值
```

```python
data.head()
```

```python
data.tail()
```

```python
data['StartTime']=data.StartTime.map(lambda x:x.split(':')[0])
```

```python
data['StartDate	']=data.StartDate.map(lambda x:x.split('/')[0])
```

```python
data.head()
```

```python
table=data.pivot_table(values='TripDurationSec',index='StartDay',columns='MemberType',aggfunc=np.mean)#data.,np.mean,pivot_table
```

```python
table
```

```python
table.plot.barh(stacked=True)
```

```python
table.sum(1)
```

```python
table_norm = table.div(table.sum(1),axis = 0)#按列逐个除sum（1），sum（1）在行上元素求和
```

```python
table_norm
```

```python
table_norm=table.div(table.sum(1),axis=0)#列之间的元素除以sum，sum(1)横着元素求和
```

```python
table_norm.plot.barh(stacked=True)#各类人群的占比
```

```python
data.pivot_table(values='TripDurationSec',index='StartTime',columns='MemberType',aggfunc=np.mean).plot.barh(figsize=(10,12),stacked=True)
```

```python
shuliang=data.EndStationName.value_counts()-data.StartStationName.value_counts()#values_counts()
```

```python
shuliang.sort_values(ascending=False)#sort_values(ascending=False),当天车站最终数量，降序
```
