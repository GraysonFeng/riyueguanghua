## 散点图

```python
df=pd.DataFrame(np.random.rand(50,4),columns=list('abcd'))
```

```python
df.plot.scatter(x='a',y='b')#散点图，要指定x=,y=
```

scatter(ax,___) 将在 ax 指定的坐标区中，而不是在当前坐标区中绘制图形。选项 ax 可以位于前面的语法中的任何输入参数组合之前。

```python
ax=df.plot.scatter(x='a',y='b',color='DarkBlue',label='Group1')
df.plot.scatter(x='c',y='d',color='DarkGreen',label='Group2',ax=ax)#将两张图合并
```

```python
df.plot.scatter(x='a',y='b',s=df['c']*200)#参数s是点的大小
```

```python
series=pd.Series(3*np.random.rand(4),index=list('abcd'),name='Series')#参数name
```

```python
df=pd.DataFrame(3*np.random.rand(4,2),index=list('abcd'),columns=list('xy'))
```
