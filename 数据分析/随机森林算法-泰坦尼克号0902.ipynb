集合算法的好处：
1、可以有效防止过拟合
2、降低噪声干扰，提高模型准确性和稳定性

### ** 随机抽取一部分特征、随机抽取一部分数据进行训练 **

```python
"""参数：
决策树的个数t
构建单棵决策树使用的特征数量"""
```

```python
"""随机森林算法
1、采样样本（放回采样）
2、采样特征
3、构建决策树
4、重复t次，构建出t棵决策树
5、综合多棵决策树的预测结果，作为随机森林的预测结果。
回归问题：平均值   分类问题：少数服从多数"""
```

```python
"""采样一部分特征的原因:
确保所有特征都可以对预测结果产生影响（影响特别大的特征会在很多时候不被抽取）"""
```

### 计算特征的重要性
#### 通过查看使用该特征减少了森林中所有树多少的不纯度，来衡量特征的重要性。
#### 它在训练后自动计算每个特征的得分，并对结果进行标准化，以使所有特征的重要性总和等于1

* n_estimators 决策树数量
* max_features 单个决策树特征数量
* min_sample_leaf  决定了叶子的数量
* n_jobs 允许使用处理器的数量，若为1就是1个，-1表示无限制

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('train.csv')
```

```python
data = data[['Survived', 'Pclass',  'Sex', 'Age', 'SibSp',
       'Parch',  'Fare',  'Embarked']]#Sibsp是否与兄妹同在，Parch是否与父母同在
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())#平均值填充缺失值
```

```python
data.fillna(0,inplace=True)
```

```python
data['Sex']=[1 if x=='male' else 0 for x in data.Sex]#列表推导式，也可用 lambda
```

```python
# 独热编码
data['p1']=np.array(data["Pclass"]==1).astype(np.int32)
data['p2']=np.array(data["Pclass"]==2).astype(np.int32)
data['p3']=np.array(data["Pclass"]==3).astype(np.int32)
```

```python
del data['Pclass']
```

```python
data['e1']=np.array(data['Embarked']=='S').astype(np.int32)
data['e2']=np.array(data['Embarked']=='C').astype(np.int32)
data['e3']=np.array(data['Embarked']=='Q').astype(np.int32)
```

```python
del data['Embarked']
```

```python
data_train = data[[ 'Sex', 'Age', 'SibSp',
       'Parch',  'Fare', 'p1','p2','p3','e1','e2','e3']].values
```

```python
# 不牵扯距离的计算，不用标准化
data_target = data['Survived'].values
```

```python
data_target = data['Survived'].values.reshape(len(data),1)
```

```python
np.shape(data_train),np.shape(data_target)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(data_train,data_target,test_size=0.2)
```

```python
x_train.shape,x_test.shape
```

```python
# 使用随机森林算法
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100,n_jobs=2)
```

```python
model.fit(x_train,y_train.ravel())#ravel使变成一列
```

```python
y_train# 已经被ravel()方法更改，更改了视图
```

```python
model.score(x_test,y_test)
```

```python
model.feature_importances_
```

```python
for futh,imp in zip([ 'Sex', 'Age', 'SibSp',
       'Parch',  'Fare', 'p1','p2','p3','e1','e2','e3'],model.feature_importances_):
    print(futh,imp)#通过迭代打印特征
```
