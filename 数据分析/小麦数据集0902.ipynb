```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('C:\\Users\\chen jia ni\\Desktop\\seeds_dataset.txt',header=None,delim_whitespace=True,names=['x1','x2','x3','x4','x5','x6','x7','y'])
```

```python
data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
data.columns
```

```python
y_pred=KMeans(n_clusters=2).fit_predict(data[['x1','x2', 'x3', 'x4', 'x5', 'x6', 'x7']])#分类
```

```python
y_pred
```

```python
y_pre=pd.Series(y_pred.astype(float))
```

```python
y_pre
```

```python
y_pre=y_pre.map({0:1,1:3})
y_pre
```

```python
data_p=pd.DataFrame({'y_pre':y_pre,'y':data.y})
data_p
```

```python
data_p['acc']=data_p.y_pre==data_p.y
```

```python
data_p
```

```python
data_p.acc.sum()/len(data_p)
```

```python
from sklearn import metrics
```

```python
metrics.calinski_harabaz_score(data[['x1','x2', 'x3', 'x4', 'x5', 'x6', 'x7']],y_pred)#评分
```
