```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = pd.DataFrame({'X':np.random.randint(1,50,100),'Y':np.random.randint(1,50,100)})
```

```python
data1
```

```python
data = pd.concat([data1+50,data1])
data
```

```python
plt.style.use('ggplot')
plt.scatter(data.X,data.Y)
```

```python
from sklearn.cluster import KMeans#cluster 聚类的算法
```

```python
y_pred = KMeans(n_clusters=2).fit_predict(data)#预测
```

```python
y_pred#预测结果
```

```python
plt.scatter(data.X,data.Y,c=y_pred)#注意这里参数c的用法
```

```python
# 量化评价聚类的效果
# 计算所有点到质心的距离最小
# 同一类越具集越好
from sklearn import metrics
metrics.calinski_harabaz_score(data,y_pred)
```

```python
plt.scatter(data.X,data.Y,c=y_pred)
```

```python
y_pred2 = KMeans(n_clusters=3).fit_predict(data)#预测
```

```python
metrics.calinski_harabaz_score(data,y_pred2)
```

```python
plt.scatter(data.X,data.Y,c=y_pred2)
```
