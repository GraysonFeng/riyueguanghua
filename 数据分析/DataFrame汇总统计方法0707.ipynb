# 层次化索引

```python
data=pd.DataFrame(np.random.randint(5,10,(6,4)),index=[list('mmmmnn'),list('abcdef')])
```

```python
data
```

```python
data.unstack()#把内层索引放到column
```

```python
data.unstack(0)#外层的索引放到column
```

```python
data=pd.DataFrame(np.random.randint(5,10,(5,6)),columns=list('abcdef'))#5行6列的自然数，取值范围5到9
```

```python
data
```

```python
data.set_index(['a','b'])#设置索引a列和b列
```

```python
data.set_index(['a','b'],drop=False)#仍保留a,b两列
```

```python
data.set_index(['a','b'],inplace=True)#在原有数据集中立即生效
```

```python
data
```

```python
data.reset_index()#索引变回数据
```

```python
data
```

```python
data.reset_index(inplace=True)
```

```python
data#已经立即生效了
```
