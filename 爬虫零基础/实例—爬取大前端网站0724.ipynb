### 获取文章代码段

```python
import requests
from lxml import etree
import csv #引入csv标准库
```

```python
pre_url = 'http://www.daqianduan.com/page/'
all_url = [pre_url + str(x) for x in range(1,73)]
```

```python
all_url
```

### 解析并保存数据

比较文章和标题xpath,识别相对的xpath

```python
/html/body/section/div[1]/div/article[2]
```

```python
/html/body/section/div[1]/div/article[2]/header/h2/a#标题的xpath,后面部分是文章标题相对的xpath
```

```python
/html/body/section/div[1]/div/article[2]/p[1]/time
```

headers里的取chrome 中的Request Headers

```python
for url in  all_url:
    response = requests.get(url, headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.186 Safari/537.36'})
    selector = etree.HTML(response.text)# 文本形式源代码
    all_article = selector.xpath('//*[starts-with(@class,"excerpt excerpt-")]')# 对all_article进行解析
    for article in all_article:
        title = article.xpath('header/h2/a/text()')[0]#提取出文本，在列表里面做切片
    for time in all_article:
        time = time.xpath('p[1]/time/text()')[0]
        item = [title,time]#获取title,time
        with open('qianduan.csv','a') as csvfile:#以追加的形式写入,保存部分
            writer = csv.writer(csvfile)#注意这里是writer ,不是write
            writer.writerow(item)
            print('正在爬取：',title)       
```
