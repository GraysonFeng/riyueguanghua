网页解析lxml库与Xpath语法

```python
import requests
```

```python
from lxml import etree #etree模块
```

```python
response = requests.get('https://www.douban.com')
```

构造选择器

```python
selector=etree.HTML(response.text)#用来解析html,用来解析response的源代码文本格式
```

```python
title=selector.xpath('//div/ul/li[1]')#切片的形式，xpath语法中是从[1]开始的
```

```python
title=selector.xpath('//div/ul/li[1]/a/text()')#在a标签下的文本
```

属性用@，''和""可以嵌套，不能同类型叠用

```python
li_3=selector.xpath('//li[@class="item-inactive"]')#从根目录开始查找，属性用@，“”“具体属性”，注意用=
```

```python
li_3=selector.xpath('//*[@class="item-inactive"]')#通过*表示任意节点
```

提取属性值

```python
selector.xpath('//li[2]/a/@href')#取第二个li,/@属性，取属性值
```

```python
li_4=selector.xpath('//li[4]')[0]#取出的是元素，不加[0]取出的是list，元素可以再用xpath
```

```python
li_4.xpath('a/text()')#li_4下a标签的text()文本
```
