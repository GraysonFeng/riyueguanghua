```python
from lxml import etree
import requests
import csv
import time
```

```python
#定义一个writecsv函数
def writecsv(item):
    with open('qfang.csv','a',encoding='utf-8') as f:#以a为追加的形式，用wt会覆盖,再定义个编码utf-8
        writer =csv.writer(f)#注意用csv.writer函数
        try:
            writer.writerow(item)#writerow按照行来写入
        except:
            print('write error')#写入失败后告知     
```

```python
if __name__ == '__main__':#当程序名字为主程序的话就执行下面的程序，否则就不执行
    headers ={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.186 Safari/537.36'}
    start_url = 'https://shanghai.qfang.com/sale/f'
    for x in range(1,5):
        url = start_url + str(x)#把range里的int转换为str
        html = requests.get(url,headers=headers)#获取了response,即url
        time.sleep(1)#防止频繁向网站请求
        selector = etree.HTML(html.text)#构造一个选择器,html的文本格式，否则是二进制
        xiaoqulist = selector.xpath('//*[@id="cycleListings"]/ul/li')#注意双引号和单引号交替使用,用xpath
        for xiaoqu in xiaoqulist:
            mingcheng = xiaoqu.xpath('div[1]/p[3]/span[2]/a[3]/text()')[0]#xpath可叠加使用
            qu = xiaoqu.xpath('div[1]/p[3]/span[2]/a[1]/text()')[0]
            quyu = xiaoqu.xpath('div[1]/p[3]/span[2]/a[2]/text()')[0]
            junjia = xiaoqu.xpath('div[2]/p/text()')[0]
            item = [mingcheng,qu,quyu,junjia]
            writecsv(item)#函数，用来保存item
            
            print('正在抓取：',mingcheng)
```
