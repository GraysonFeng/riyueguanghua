```python
from lxml import etree
import requests
import csv
from multiprocessing.dummy import Pool as pl     #导入线程池
```

```python
def towrite(item):
    with open('qfang_xiaoqu.csv', 'a',encoding = 'utf-8') as csvfile:
        writer = csv.writer(csvfile)
        try:
            writer.writerow(item)
        except:
            print('write error!')
```

```python
#定义的方法，在同一URL下做的事
def spider(url):
    htm = requests.get(url, headers=headers)
    response=etree.HTML(htm.text)
    mingcheng = response.xpath('//*[@id="headInfo"]/div/div[2]/div[2]/ul/li[1]/p/a/text()')[0]
    xiaoqu_item = [mingcheng]
    towrite(xiaoqu_item)
    print('正在爬取小区：',mingcheng)
    tupian_url = response.xpath('//*[@id="hsPics"]/ul/li[1]/a/img/@src')[0]#图片位置和案例不同，所以要自取图片网页地址
    b_tupian = requests.get(tupian_url,headers=headers)#用前面的headers
    with open('./xiaoqutupian/' + mingcheng + '.jpg','wb') as tp:#需要现在同一文件夹目录下创建目录
        tp.write(b_tupian.content)
        print('正在下载图片：',mingcheng)
```

```python
#定义的是URL
if  __name__ == '__main__': 
    headers = {
       'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.75 Safari/537.36'}
    start_url = 'https://shanghai.qfang.com/sale/f'
    pool = pl()
    all_url = []
    for x in range(1,5):
        html = requests.get(start_url + str(x), headers=headers)
        slector = etree.HTML(html.text)
        xiaoqulist = slector.xpath('//*[@id="cycleListings"]/ul/li')
        for xiaoqu in xiaoqulist:
            xiaoqu_url_houduan = xiaoqu.xpath('div[1]/p[1]/a/@href')[0]#通过首页链接找href 关联网页地址
            xiaoqu_url = 'https://shanghai.qfang.com' + xiaoqu_url_houduan
            all_url.append(xiaoqu_url)
    pool.map(spider,all_url)
    pool.close()#关闭进程池，不接受任务了
    pool.join()#等待所有线程结束，收集所有的线程
```
