## 定义网络

```python
x = tf.placeholder("float",shape=[None,11])
y = tf.placeholder("float",shape=[None,1])
```

weight权重输入值的行数与x的列数对应，输出值是一个标量

```python
weight = tf.Variable(tf.random_normal([11,1]))#权重
bias = tf.Variable(tf.random_normal([1]))#偏置
output = tf.matmul(x,weight) + bias #matmul矩阵乘法
pred = tf.cast(tf.sigmoid(output)>0.5,tf.float32) #cast改变数值类型，计算概率值sigmoid大于0.5
```

```python
loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels = y,logits=output))#tf.reduce_mean沿着某一维度计算元素的平均值
```

```python
train_step = tf.train.GradientDescentOptimizer(0.001).minimize(loss)#梯度下降法,最小化损失函数
```

```python
accuracy = tf.reduce_mean(tf.cast(tf.equal(pred,y),tf.float32))#布尔值转换为数值，求平均值
```

进行训练

```python
sess = tf.Session()#初始化创建会话
sess.run(tf.global_variables_initializer())#初始化一个变量
```

```python
for i in range(10000):#训练1万步
    for n in range(len(data_target)//100):#每100行输入以下值打乱训练数据，增加泛化能力,用目标数据的长度//整除,
        index = np.random.permutation(len(data_target))#乱序
        data_data = data_data[index]#改变data_data顺序
        data_target = data_target[index]#增加泛化能力
        batch_xs = data_data[n:n+100]#每次取出100个值
        batch_ys =data_target[n:n+100]
        sess.run(train_step,feed_dict={x:batch_xs,y:batch_ys})#feed_dict训练
    if i%1000 == 0:
        print(sess.run((loss,accuracy),feed_dict={x:batch_xs,y:batch_ys}))#每隔1000步看loss,accuracy
```
